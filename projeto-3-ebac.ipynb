## Análise exploratória de dados de crédito com SQL

Esse projeto faz parte de uma tarefa prática do curso Profissão: Analista de Dados da EBAC da parte de SQL. 
O objetivo do projeto é fazer uma análise exploratória de dados sobre uma planilha com dados de crédito, como a planilha original contém aproximadamente 10000 linhas o conjunto de dados foi reduzido para um total de 2564 linhas no projeto.

**Descrição dos dados:**

- Idade: Idade do cliente.
- Sexo: Sexo do cliente ("M" para masculino e "F" para feminino).
- Dependentes: Número de filhos.
- Escolaridade: Nível escolar.
- Estado_civil: Se é solteiro, casado e etc...
- Salario_anual: Faixa de salário ganha ao ano.
- Tipo_cartao: Tipo de cartão do cliente.
- Qtd_produtos: Quantidade de produtos comprados.
- Iteracoes_12m: Número de interações nos últimos 12 meses.
- Meses_inativo_12m: Número de meses inativo nos últimos 12 meses.
- Limite_credito: Valor do limite do cartão de crédito.
- Valor_transacoes_12m: Valor das transações nos últimos 12 meses.
- Qtd_transacoes_12m: Número de transações nos últimos 12 meses.

## Análise de Dados

Na primeira consulta na parte de análise de dados temos a distribuição dos clientes quanto a sua faixa salarial, vemos que a faixa salarial com mais clientes é a de menos de 40 mil, porém a grande maioria dos clientes com informações disponíveis recebe acima de 40 mil e não há informações quanto a faixa salarial de 235 clientes.

Query 5: select count(*), salario_anual from credito8 group by salario_anual

![Faixa Salarial](https://github.com/guerrinha77/Imagens/blob/main/Captura%20de%20tela%202025-11-06%20181944.png?raw=true)

Na query seguinte temos o número de clientes divididos por sexo. Os homens são maioria.

Query 6: select count(*), sexo from credito8 group by sexo

![Sexo](https://github.com/guerrinha77/Imagens/blob/main/Captura%20de%20tela%202025-11-06%20182023.png?raw=true)

Nessa nova consulta filtramos os dados de maneira que obtemos os 10 clientes com maior limite de crédito, vemos a escolaridade, o tipo de cartão e o seu sexo. Surpreendentemente o 1º e o 4º não possuem educação formal e apenas duas mulheres figuram entre os 10 maiores limite de crédito.

Query 7: select max(limite_credito) as limite_credito, escolaridade, tipo_cartao, sexo from credito8 where escolaridade != 'na' and tipo_cartao != 'na' group by escolaridade, tipo_cartao, sexo order by limite_credito desc limit 10

![Top 10 limites de crédito](https://github.com/guerrinha77/Imagens/blob/main/Captura%20de%20tela%202025-11-06%20182054.png?raw=true)

Na query abaixo temos uma comparação entre mulheres e homens quanto ao gastos, destacando a media, maior e menor valor gasto. Mulheres possuem o maior e menor valor gasto em relação aos homens assim como uma media de gastos maior.

Query 8: select max(valor_transacoes_12m) as maior_valor_gasto, avg(valor_transacoes_12m) as media_valor_gasto, min(valor_transacoes_12m) as min_valor_gasto, sexo from credito8 group by sexo

![Gastos Mulheres vs Homens](https://github.com/guerrinha77/Imagens/blob/main/Captura%20de%20tela%202025-11-06%20182123.png?raw=true)

Pegando agora médias da quantidade de produtos, valor de transações e limite de crédito juntamente com o sexo e a faixa salarial.

Query 9: select avg(qtd_produtos) as qts_produtos, avg(valor_transacoes_12m) as media_valor_transacoes, avg(limite_credito) as media_limite, sexo, salario_anual from credito8 where salario_anual != 'na' group by sexo, salario_anual order by avg(valor_transacoes_12m) desc

![Médias](https://github.com/guerrinha77/Imagens/blob/main/Captura%20de%20tela%202025-11-06%20182151.png?raw=true)

## Conclusão

Insights obtidos apartir do conjunto de dados:

- A maior parte dos cliente são homens.
- A maior parte dos clientes possuem salário anual maior que 40 mil reais.
- O cliente com maior limite de crédito é homem e não possui educação formal.
- As mulheres tem um gasto médio maior que os homens no geral.
